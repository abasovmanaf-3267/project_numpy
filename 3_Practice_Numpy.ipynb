```python
### 7  1-dən 20-yə qədər rəqəmlərdən ibarət array yaradın, sonra yalnız 5-ə və ya 7-yə bölünən rəqəmləri saxlayaraq yeni massiv yaradın.

arr = np.arange(1, 21)
filtered = arr[(arr % 5 == 0) | (arr % 7 == 0)]
print(filtered)
# Çıxış: [ 5  7 10 14 15 20]
```
